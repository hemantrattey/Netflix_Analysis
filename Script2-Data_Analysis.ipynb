```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pwd = os.getcwd()
pwd
```

```python
filepath = os.path.join(pwd, 'final_data.csv')
filepath
```

```python
df_import = pd.read_csv(filepath)
df_import.head()
```

```python
df = df_import.copy()
```

```python
df.info()
```

```python
# temp = '2021-04-19 16:40:06+05:30'.split('+')[0].split(' ')[1]
# temp
```

```python
# extract timestamp as a seperate column 
df['Start Time'] =  df['Date'].apply(lambda x : x.split('+')[0].split(' ')[1])
```

```python
df.head()
```

```python
# Remove timestamp from date 
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.date
```

```python
df.head()
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
# Convert duration HH:MM:SS to number of minutes 
df['Duration (Seconds)'] = df['Duration'].str.split(':').apply(lambda x: int(x[0])*3600 + int(x[1])*60 + int(x[2]))
```

```python
df.head()
```

```python
df.info()
```

```python
avg_seconds_per_day = df.groupby('day_name')['Duration (Seconds)'].mean().sort_values()
avg_seconds_per_day
```

```python
type(avg_seconds_per_day)
```

```python
day_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
avg_seconds_per_day.values
```

```python
sns.set_style("darkgrid")
font = {'family': 'serif',
        'color': '#004466',
        'weight': 'normal',
        'size': 16}
plt.figure(figsize=(20,15))
ax = sns.barplot(x = avg_seconds_per_day.index, y = avg_seconds_per_day.values, palette=("Blues_d"), ci = None, order=day_of_week)
ax.set_ylabel('Viewing Duration (in Seconds)',fontdict={'size': 13, 'family': 'serif'})
ax.set_xlabel('Day of Week', fontdict={'size': 13, 'family': 'serif'})
ax.set_title('Average Viewing Time per Day (in seconds)', fontdict=font)
ax.tick_params(axis = 'both', labelsize= 12)
plt.savefig(os.path.join(pwd, 'viz', 'Average Viewing Duration per Day (seconds).png'))
```

```python
tv_shows = df[df['Content Type'] == 'TV Show']
movies = df[df['Content Type'] == 'Movie']
```

```python
tv_shows.head()
```

```python
tv_shows[tv_shows['Episode'].isna()]
```

```python
tv_shows['Episode'].fillna(tv_shows['Season'], inplace = True)
```

```python
tv_shows['Title'].nunique()
```

```python
most_watched_tv_series = tv_shows.groupby('TV Show')['Duration (Seconds)'].mean().reset_index().sort_values(by = 'Duration (Seconds)', ascending = False)
most_watched_tv_series.head(10)
```

```python
font = {'family': 'serif',
        'color': '#004466',
        'weight': 'normal',
        'size': 16}
plt.figure(figsize=(22, 12))
ax = sns.barplot(y = most_watched_tv_series['TV Show'][:10], x = most_watched_tv_series['Duration (Seconds)'][:10], orient = 'h', ci = 'None')
ax.set_ylabel('TV Series', fontdict={'size': 13, 'family': 'serif'})
ax.set_xlabel('Watch Time (in seconds)', fontdict={'size': 13, 'family': 'serif'})
ax.set_title('Top 10 TV Shows by Average Duration (seconds)', fontdict=font)
ax.tick_params(axis = 'both', labelsize= 12)
plt.savefig(os.path.join(pwd, 'viz', 'Top 10 TV Shows by Average Duration (seconds).png'))
```

```python
# Counting episodes by TV SHOW and Date would also count episodes when I watched only 1 second and quit the app or tab. Thus a better way would be to filter out rows where duration < 5 mins or whatever feels right. But I won't do this here as I have limited amount of data. But this should be kept in mind if you have a good amount of data. That will lead to a better analysis 

binge = tv_shows.groupby([tv_shows.index, 'TV Show'])['Episode'].count().reset_index().sort_values(by='Episode', ascending=False)
binge
```

```python
# Filter and remove TV Shows where episodes watched in a day < 5. I will consider binge watching if episodes by TV show per day is greater than 5. 
 
binge = binge[binge['Episode'] > 5]
binge
```

```python
# Group all the TV shows and sum the episodes. Use this to plot the top TV shows by number of episodes binged. 
top_10_binge = binge.groupby('TV Show')['Episode'].sum().reset_index().sort_values(by = 'Episode', ascending = False)
top_10_binge
```

```python
font = {'family': 'serif',
        'color': '#004466',
        'weight': 'normal',
        'size': 16}
plt.figure(figsize=(22, 12))
ax = sns.barplot(y = top_10_binge['TV Show'][:10], x = top_10_binge['Episode'][:10], orient = 'h', ci = 'None')
ax.set_ylabel('TV Series', fontdict={'size': 13, 'family': 'serif'})
ax.set_xlabel('Number of Episodes', fontdict={'size': 13, 'family': 'serif'})
ax.set_title('Top 10 Binge Watching TV Series', fontdict=font)
ax.tick_params(axis = 'both', labelsize= 12)
plt.savefig(os.path.join(pwd, 'viz', 'Top 10 Binge Watching TV Series.png'))
```
